# DATA PREPARATION

Data preparation will involve; importing neccessary modules ,data previews for better data understanding,data cleaning(dealing with missing data) and handling non_numerically data for modeling.

```python
#importing neccessary modules
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data preview
data=pd.read_csv('kc_house_data.csv',index_col=0)
data.head()
```

```python
data.info()
```

```python
data_categoricals = data.select_dtypes("object")

for col in data_categoricals:
    print(data[col].value_counts(), "\n")
```

```python
#handling missing data for columns we might be intrested in

data['waterfront'] = data['waterfront'].fillna('NO')
data['view'] = data['view'].fillna('NONE')
```

```python
#checking for numeric column most correlated to price
data_numeric = data.select_dtypes(include=['int','float']).corr()['price']
data_numeric
```

```python
#getting a list of columns with high correration to price corr>0.5
price_high_corr=data_numeric[data_numeric>0.5].index.tolist()
price_high_corr
```

To avoid **multicollinearity** we need to check the correlation between our highly price correlated predictors.WE will use a define high correlation as one above 0.75 in this project.

```python
#checking for correlation between our predictors
price_high_corr_data=data[price_high_corr]
price_high_corr_data.corr()
```

All the selected numeric predictors are highly correlated to **sqft_living**.To avoid multicollinearity we will just use the sqft_living column among the numeric data to predict the price.

Next we will look at the categorical data.The categorical data that probably have an impact on the price on our data are
waterfront,condition,view and grade.We will first plot bar graphs of each of this predictor to have an idea of the predictor impact on price and a more understanding of the data.

```python
#bar plot for categorical data
fig, axes = plt.subplots(ncols=4, figsize=(12,5))
data.groupby('waterfront')['price'].mean().plot.bar(y="price", ax=axes[0])
data.groupby('condition')['price'].mean().plot.bar(y="price", ax=axes[1])
data.groupby('grade')['price'].mean().plot.bar(y="price", ax=axes[3])
data.groupby('view')['price'].mean().plot.bar(y="price", ax=axes[2]);
```

From the plotted bar graphs all the four categorical data sets seems to influence the prices of the houses.House with a waterfront look more expensive than those with none.houses with good to very good condition seems to have fetched better prices.houses with excellent and those with good views look expensive while houses of good grades reflects very high prices compared to those of lower grades

With these consinderations we are left with 6 columns of intrest.price the dependent valiable and 5 predictors(sqft_living,view,waterfront ,condition and grade)

```python
#filtering our data to our columns of intrest
selected_cols=['price','sqft_living','view','waterfront','condition','grade']
selected_data=data[selected_cols]
selected_data
```

```python
#Transforming Categorical Variables
selected_cols_dum = pd.get_dummies(selected_data, columns=["view",'waterfront','condition','grade']).astype(int)
selected_cols_dum = selected_cols_dum.drop(["view_NONE",'waterfront_NO','condition_Average','grade_7 Average'], axis=1)
selected_cols_dum
```

# MODELING

```python
#scatter plot to view the distribution of sqft_living against price
data.plot.scatter(x='sqft_living', y="price");
```

```python

y=selected_data['price']
X=selected_data['sqft_living']
baseline_model = sm.OLS(y, sm.add_constant(X))
baseline_results = baseline_model.fit()
```

**Model Evaluation**
In every model that we make we will need to evaluate the model by checking its parameters,R_squared and the p values of the model and its parameters.We can get this by using .summarry but it returns a wide range of results which we sometimes dont need.we will create a small function to return only the information we need incases that we find .summary output to be bulky.

```python
#create a small function to return only the information we need incases that we find .summary output to be bulky.

def model_summary(model):
    return print(f'''model R squared value \n {model.rsquared}\n \n model p_value \n {model.f_pvalue} \n 
    model parameters \n {model.params}\n parameters p_values \n {model.pvalues}''')

#testing the function with our first model
model_summary(baseline_results)
```

**Model Interpretation**

This translates to our model formula being,
**price=281(sqft_living)-43989**


Our first simple regression model has R_squared of about 0.49. This means 49% of the variation in price is explained by our model.
Our model has an overall p_value of 0 which is below the typical alpha of 0.05.This mean that our model is statistically significant.

Our model has slope of 281 which mean an increase of sqft_living by 1 increase the price by 281 units.
the y intercept is -43989 which is means when the living area is zero the price is -43989.That does not make sense but we will transform our more advanced models for an interpretable y intercept.


```python
#a view of our data points with the regression line
#plotting scatter
fig, ax = plt.subplots();
plt.scatter(selected_data['sqft_living'],selected_data['price'],label='data points')
#plotting the regresssion line
price=281*(selected_data['sqft_living'])-43989
ax.plot(selected_data['sqft_living'], price, label="Regression Line",color='red')
#labeling
ax.set_xlabel('sqft_living')
ax.set_ylabel('price')
fig.suptitle("Scatter Plot sqft_living vs price with regression line")
ax.legend();
```

Using **Polynomial Regression** to improving our r_squared therefore improving our variation explanation.

```python
#Creating a 2nd-degree polynomial (squared term)
selected_data_sq=selected_data.copy()
selected_data_sq['squared']=selected_data_sq['sqft_living']**2
#creating the model
y=selected_data['price']
X_sq=selected_data_sq[['sqft_living','squared']]
baseline_model_sq = sm.OLS(y, sm.add_constant(X_sq))
baseline_results_sq = baseline_model_sq.fit()
#evaluating our model
model_summary(baseline_results_sq)
```

```python
#a view of our data points with the Polynomial regression line
#plotting scatter
fig, ax = plt.subplots();
plt.scatter(selected_data['sqft_living'],selected_data['price'],label='data points')
#plotting the Polynomial regresssion line
price=0.038609*(selected_data['sqft_living'])**2+86*(selected_data['sqft_living'])+199186
ax.plot(selected_data['sqft_living'], price, label="Regression Line",color='red')
#labeling
ax.set_xlabel('sqft_living_squared')
ax.set_ylabel('price')
fig.suptitle("Scatter Plot sqft_living_squared vs price with regression line")
ax.legend();
```

**Adding our categorical data to our model**

```python
#creating model from our prepared data
X_selected_cols_dum=selected_cols_dum.drop(["price"], axis=1)
model_X_selected_cols_dum = sm.OLS(y, sm.add_constant(X_selected_cols_dum))
results_X_selected_cols_dum = model_X_selected_cols_dum.fit()
#checking for r_squared and f_pvalue
print(results_X_selected_cols_dum.rsquared)
print(results_X_selected_cols_dum.f_pvalue)
```

```python
#improving our r_squared by squaring the sqft_living
X_selected_cols_dum_sq=X_selected_cols_dum**2
model_X_selected_cols_dum_sq = sm.OLS(y, sm.add_constant(X_selected_cols_dum_sq))
results_X_selected_cols_dum_sq = model_X_selected_cols_dum_sq.fit()
results_X_selected_cols_dum_sq.summary()
print(results_X_selected_cols_dum_sq.rsquared)
print(results_X_selected_cols_dum_sq.f_pvalue)
```

Adding our categorical data and squaring of our numerical data has improved our model variance explanation from around 49% to about 65%.our p value is zero therefore our model is still statistically viable.

Next we will transform our model to have a better y intercept which will be price of an average sized house.

```python
#transform our model to have a better y intercept
X_selected_cols_dum_sq['sqft_living']=X_selected_cols_dum_sq['sqft_living']-np.mean(X_selected_cols_dum_sq['sqft_living'])
X_selected_cols_dum_sq['sqft_living']
model_X_selected_cols_dum_sq = sm.OLS(y, sm.add_constant(X_selected_cols_dum_sq))
results_X_selected_cols_dum_sq = model_X_selected_cols_dum_sq.fit()
#final model summary
results_X_selected_cols_dum_sq.summary()
```
